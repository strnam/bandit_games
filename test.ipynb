```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Draw 1000 samples from Beta(1,1)
prior_a = 1
prior_b = 2
samples = beta.rvs(prior_a, prior_b, size=1000)

# Plot histogram and theoretical PDF
x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, prior_a, prior_b)

plt.hist(samples, bins=20, density=True, alpha=0.6, label='Sample Histogram')
plt.plot(x, pdf, 'r-', lw=2, label=f'Beta({prior_a},{prior_b}) PDF')
plt.title(f'Beta({prior_a},{prior_b}) Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Draw 1000 samples from Beta(1,1)
samples = beta.rvs(10, 1, size=1000)

# Plot histogram and theoretical PDF
x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, 10, 1)

plt.hist(samples, bins=20, density=True, alpha=0.6, label='Sample Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='Beta(1,1) PDF')
plt.title('Beta(1,1) Distribution (Uniform[0,1])')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
